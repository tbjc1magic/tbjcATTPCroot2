```python
from sqlalchemy import *
import numpy as np
import pylab as plt
import pandas as pd
from peakdetect import peakdetect
from scipy.ndimage.interpolation import shift
from matplotlib.colors import LogNorm

%matplotlib inline
```

```python
engine = create_engine('sqlite+pysqlite:///../SQL/72_8.db')
```

```python
import time
start_time = time.time()
ADCdf = pd.io.sql.read_sql("SELECT * FROM ADC", engine)
Hitdf = pd.io.sql.read_sql("SELECT * FROM Hit", engine)
PSAdf = pd.io.sql.read_sql("SELECT * FROM PSA", engine)
ProtoMapdf = pd.io.sql.read_sql("SELECT * FROM ProtoMap", engine)
#RawADCdf = pd.io.sql.read_sql("SELECT * FROM RawADC", engine)
end_time = time.time()
print end_time-start_time
```

```python
def Process(Event):
    mesh = Event.iloc[:,3:][Event.iloc[:,3:]>20].sum(axis=0)
    #print mesh
    return mesh
start_time = time.time()
mesh_df = ADCdf.groupby('EventID').apply(Process)
end_time = time.time()
print end_time-start_time
```

```python
def convolve(event):
    mesh_convol = np.convolve(event.values,([1]*10)[::-1],'same')
    return pd.Series(mesh_convol,index=['t'+str(i) for i in range(512)])
convol_df=mesh_df.apply(convolve,axis=1)
convol_df.reset_index(inplace=True)
```

```python
status_df = pd.read_csv("72_8_t0.csv")
status_df.columns = ['EventID','trigger']
```

```python
grad = np.gradient(convol_df.iloc[0].values)
mx = np.argmax(grad)
my = grad[mx]
plt.plot(convol_df.iloc[0].values*0.1)
plt.plot(grad)
plt.plot(mx,my,'+',c='r')
```

```python
def FindLeadingEdge(e):
    
    event= e.iloc[1:-1]
    grad = np.gradient(event.values)
    mx = np.argmax(grad)
    my = grad[mx]
    
    tmp = e[['EventID','trigger']].values.tolist() +[mx,my]
    return pd.Series(tmp,index=['EventID','trigger','pos','height'])

    
tmp_df = pd.merge(convol_df,status_df,on='EventID')
tmp_df = tmp_df[~tmp_df['trigger'].isnull()]
tmp_df = tmp_df[tmp_df['trigger']<425]
peakdf = tmp_df.apply(FindLeadingEdge,axis=1)
#peakdf['pos'].hist(bins=200)
(peakdf['pos']-peakdf['trigger']+400).hist(bins=200)
plt.xlim([200,260])
plt.figure()
peakdf['pos'].hist(bins=200)
plt.xlim([180,250])
```

```python
def Shift(r):
    r1 = r.copy()
    r1.iloc[1:-1] = r1.iloc[1:-1].shift(int(380-r1['trigger']))
    return  r1.fillna(0)
tmp_df = pd.merge(convol_df,status_df,on='EventID')
tmp_df = tmp_df[~(tmp_df['trigger'].isnull())]
shift_df = tmp_df.apply(Shift,axis=1)
shift_df.iloc[1:-1].T.plot(c=(1,0,0,0.01),legend=None)
```

```python
%matplotlib notebook
heights = shift_df.iloc[:,1:-1].max(axis=1)
pos = shift_df.apply(FindLeadingEdge,axis=1)

#mask = (pos['pos']>205)&(pos['pos']<230)&(heights>42000)&(heights<53000)
mask = (pos['pos']>180)&(pos['pos']<220)&(heights>53000)&(heights<65000)

#heights[mask].hist()
#plt.figure()
print pos['pos'][mask].shape
plt.hist2d(pos['pos'].values,heights.values,bins=500,norm=LogNorm());
#plt.hist2d(pos['pos'][mask].values,heights[mask].values,bins=[np.linspace(200,230,15),np.linspace(44000,52000,20)],norm=LogNorm());
#plt.xlim([150,250]);
#plt.ylim([40000,80000]);
```
