```python
import ROOT
import numpy as np
import pylab as plt
from array import array

%matplotlib inline
```

```python
ROOT.gROOT.ProcessLine(
"struct tbjcStruct\
{\
    short uval[512];\
};" )
```

```python
f = ROOT.TFile("../data/ANL_March_2013/ROOT_files/run28.root")
tree = f.Get("AutoTree")
tb0 = tree.GetBranch("asic_0_00")
#tb1 = tree.GetBranch("PSA")
#tb2 = tree.GetBranch("Hit")
#l = ROOT.vector('short')()
tbjcStruct = ROOT.tbjcStruct()
tb0.SetAddress(ROOT.AddressOf(tbjcStruct,'uval'))
```
